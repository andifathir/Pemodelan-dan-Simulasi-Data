```python
import numpy as np
import matplotlib.pyplot as plt


initial_temperature = 100
ambient_temperature = 25
time_steps = 200
dt = 0.1

cooling_rates = [0.05, 0.1, 0.15]

time = np.arange(0, time_steps * dt, dt)

plt.figure(figsize=(8, 6))

for cooling_rate in cooling_rates:
    temperature = np.zeros(len(time))
    temperature[0] = initial_temperature
    for t in range(1, len(time)):
        dT = -cooling_rate * (temperature[t-1] - ambient_temperature) * dt
        temperature[t] = temperature[t-1] + dT
    
    plt.plot(time, temperature, label=f'Cooling Rate = {cooling_rate}')

plt.xlabel("Time (s)")
plt.ylabel("Temperature (°C)")
plt.title("Effect of Cooling Rate on Temperature Decay")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('weatherHistory.csv')

temperature_data = data[['Formatted Date', 'Temperature (C)']]

temperature_data.loc[:, 'Formatted Date'] = pd.to_datetime(temperature_data['Formatted Date'], utc=True)  # Ensure UTC time

cooling_rate = 0.05
dt = 0.1

ambient_temperature = temperature_data['Temperature (C)'].values

time = np.arange(0, len(ambient_temperature) * dt, dt)
temperature = np.zeros(len(time))

initial_temperature = ambient_temperature[0]
temperature[0] = initial_temperature

for t in range(1, len(time)):
    current_ambient_temp = ambient_temperature[t % len(ambient_temperature)]
    dT = -cooling_rate * (temperature[t-1] - current_ambient_temp) * dt
    temperature[t] = temperature[t-1] + dT

plt.plot(time, temperature, label="Simulated Object Temperature (Based on Weather Data)")
plt.xlabel("Time (s)")
plt.ylabel("Temperature (°C)")
plt.title("Temperature Changes Simulated from Weather Data")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

initial_temperature = 100
ambient_temperature = 25 
cooling_rate = 0.05
time_steps = 200
dt = 0.1

time = np.arange(0, time_steps * dt, dt)
temperature = np.zeros(len(time))
temperature[0] = initial_temperature

# dT/dt = -k (T - T_ambient)
for t in range(1, len(time)):
    dT = -cooling_rate * (temperature[t-1] - ambient_temperature) * dt
    temperature[t] = temperature[t-1] + dT

plt.plot(time, temperature, label="Object Temperature")
plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.title("Continuous Event Simulation (CES): Newton's Law of Cooling")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

initial_temperature = 100
ambient_temperature = 25
cooling_rate = 0.05
time_steps = 200
event_interval = 10

time = np.arange(0, time_steps)
temperature = initial_temperature
temperature_history = [temperature]

for t in range(1, len(time)):
    if temperature < 30 and t % event_interval == 0:
        temperature += 2
    
    dT = -cooling_rate * (temperature - ambient_temperature)
    temperature += dT

    temperature_history.append(temperature)

plt.plot(time, temperature_history, label="Object Temperature")
plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.title("Discrete Event Simulation (DES): Heat Transfer with Events")
plt.legend()
plt.grid()
plt.show()
```
