```python
import random
import math

# Fungsi untuk mensimulasikan Monte Carlo untuk menghitung pi
def monte_carlo_pi(num_simulations, a):
    # Counter untuk bola di masing-masing mangkuk
    square_bowl_count = 0
    circular_bowl_count = 0
    
    for _ in range(num_simulations):
        # Tempatkan bola secara acak di dalam persegi (dalam rentang 0 hingga a)
        x = random.uniform(0, a)
        y = random.uniform(0, a)
        
        # Hitung apakah bola jatuh di dalam persegi
        if x <= a and y <= a:
            square_bowl_count += 1
            # Periksa apakah bola jatuh dalam lingkaran
            if x**2 + y**2 <= a**2:
                circular_bowl_count += 1
    
    # Rasio bola di mangkuk lingkaran terhadap bola di mangkuk persegi
    pi_estimate = (circular_bowl_count / square_bowl_count) * 4
    return pi_estimate

# Jumlah simulasi dan ukuran sisi persegi
num_simulations = 10000
a = 1  # Sisi persegi dan jari-jari lingkaran adalah 1 (bisa diganti)

# Estimasi pi
estimated_pi = monte_carlo_pi(num_simulations, a)
print(f"Estimasi nilai pi: {estimated_pi}")
```

```python
import random
import math
import matplotlib.pyplot as plt

# Fungsi untuk mensimulasikan Monte Carlo untuk menghitung pi dan menggambar visualisasi
def monte_carlo_pi(num_simulations, a):
    # Counter untuk bola di masing-masing mangkuk
    square_bowl_count = 0
    circular_bowl_count = 0
    
    # List untuk menyimpan koordinat bola yang jatuh
    x_inside_circle = []
    y_inside_circle = []
    x_outside_circle = []
    y_outside_circle = []
    
    for _ in range(num_simulations):
        # Tempatkan bola secara acak di dalam persegi (dalam rentang 0 hingga a)
        x = random.uniform(0, a)
        y = random.uniform(0, a)
        
        # Hitung apakah bola jatuh di dalam persegi
        square_bowl_count += 1
        # Periksa apakah bola jatuh dalam lingkaran
        if x**2 + y**2 <= a**2:
            circular_bowl_count += 1
            # Simpan koordinat bola yang jatuh dalam lingkaran
            x_inside_circle.append(x)
            y_inside_circle.append(y)
        else:
            # Simpan koordinat bola yang jatuh di luar lingkaran
            x_outside_circle.append(x)
            y_outside_circle.append(y)
    
    # Rasio bola di mangkuk lingkaran terhadap bola di mangkuk persegi
    pi_estimate = (circular_bowl_count / square_bowl_count) * 4
    
    # Visualisasi
    plt.figure(figsize=(6, 6))
    
    # Gambar persegi
    plt.plot([0, a, a, 0, 0], [0, 0, a, a, 0], color='black', linewidth=2, label='Square Bowl')
    
    # Gambar lingkaran
    circle = plt.Circle((a/2, a/2), a/2, color='blue', fill=False, linestyle='--', label='Circular Bowl')
    plt.gca().add_artist(circle)
    
    # Gambar bola yang jatuh dalam lingkaran
    plt.scatter(x_inside_circle, y_inside_circle, color='green', s=1, label='Inside Circle')
    
    # Gambar bola yang jatuh di luar lingkaran
    plt.scatter(x_outside_circle, y_outside_circle, color='red', s=1, label='Outside Circle')
    
    plt.xlim(0, a)
    plt.ylim(0, a)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(f"Monte Carlo Simulation for Estimating Pi\nEstimated Pi: {pi_estimate}")
    plt.legend()
    plt.show()
    
    return pi_estimate

# Jumlah simulasi dan ukuran sisi persegi
num_simulations = 10000
a = 1  # Sisi persegi dan jari-jari lingkaran adalah 1 (bisa diganti)

# Estimasi pi dan visualisasi
estimated_pi = monte_carlo_pi(num_simulations, a)
print(f"Estimasi nilai pi: {estimated_pi}")
```

```python
import random
import math

# Fungsi untuk mensimulasikan Monte Carlo untuk menghitung pi
def monte_carlo_pi(num_simulations, a, table_size):
    # Counter untuk bola di masing-masing mangkuk
    square_bowl_count = 0
    circular_bowl_count = 0
    no_bowl_count = 0  # Bola yang tidak masuk ke dalam mangkuk
    
    for _ in range(num_simulations):
        # Tempatkan bola secara acak pada meja, dalam rentang 0 hingga table_size (meja lebih besar dari mangkuk)
        x = random.uniform(0, table_size)
        y = random.uniform(0, table_size)
        
        # Periksa apakah bola jatuh di dalam persegi (square bowl)
        if x <= a and y <= a:
            square_bowl_count += 1
            # Periksa apakah bola juga jatuh di dalam lingkaran (circular bowl)
            if x**2 + y**2 <= a**2:
                circular_bowl_count += 1
        # Jika bola tidak masuk ke dalam kedua mangkuk (mungkin jatuh di luar mangkuk)
        elif x > a or y > a:
            no_bowl_count += 1
    
    # Rasio bola di mangkuk lingkaran terhadap bola di mangkuk persegi
    pi_estimate = (circular_bowl_count / square_bowl_count) * 4
    return pi_estimate, square_bowl_count, circular_bowl_count, no_bowl_count

# Jumlah simulasi dan ukuran sisi persegi
num_simulations = 10000
a = 1  # Sisi persegi dan jari-jari lingkaran adalah 1 (bisa diganti)
table_size = 3  # Ukuran meja lebih besar dari mangkuk (misalnya, 3x3)

# Estimasi pi
estimated_pi, square_count, circle_count, no_bowl_count = monte_carlo_pi(num_simulations, a, table_size)

# Menampilkan hasil
print(f"Estimasi nilai pi: {estimated_pi}")
print(f"Bola di mangkuk persegi: {square_count}")
print(f"Bola di mangkuk lingkaran: {circle_count}")
print(f"Bola yang tidak masuk ke kedua mangkuk: {no_bowl_count}")
```

```python
import random
import math
import matplotlib.pyplot as plt

# Fungsi untuk mensimulasikan Monte Carlo untuk menghitung pi
def monte_carlo_pi(num_simulations, a, table_size):
    # Counter untuk bola di masing-masing mangkuk
    square_bowl_count = 0
    circular_bowl_count = 0
    no_bowl_count = 0  # Bola yang tidak masuk ke dalam mangkuk
    inside_square_x = []
    inside_square_y = []
    inside_circle_x = []
    inside_circle_y = []
    outside_bowls_x = []
    outside_bowls_y = []

    for _ in range(num_simulations):
        # Tempatkan bola secara acak pada meja, dalam rentang 0 hingga table_size
        x = random.uniform(0, table_size)
        y = random.uniform(0, table_size)

        # Jika bola jatuh dalam persegi
        if x <= a and y <= a:
            inside_square_x.append(x)
            inside_square_y.append(y)
            square_bowl_count += 1

            # Periksa apakah bola juga jatuh di dalam lingkaran
            if x**2 + y**2 <= a**2:
                inside_circle_x.append(x)
                inside_circle_y.append(y)
                circular_bowl_count += 1
        # Jika bola jatuh di luar kedua mangkuk
        else:
            outside_bowls_x.append(x)
            outside_bowls_y.append(y)
            no_bowl_count += 1

    # Rasio bola di mangkuk lingkaran terhadap bola di mangkuk persegi
    pi_estimate = (circular_bowl_count / square_bowl_count) * 4
    return pi_estimate, inside_square_x, inside_square_y, inside_circle_x, inside_circle_y, outside_bowls_x, outside_bowls_y

# Jumlah simulasi dan ukuran sisi persegi
num_simulations = 10000
a = 1  # Sisi persegi dan jari-jari lingkaran adalah 1 (bisa diganti)
table_size = 3  # Ukuran meja lebih besar dari mangkuk (misalnya, 3x3)

# Estimasi pi
estimated_pi, inside_square_x, inside_square_y, inside_circle_x, inside_circle_y, outside_bowls_x, outside_bowls_y = monte_carlo_pi(num_simulations, a, table_size)

# Menampilkan hasil estimasi pi
print(f"Estimasi nilai pi: {estimated_pi}")

# Visualisasi
fig, ax = plt.subplots()

# Plot meja (seluruh area)
ax.set_xlim(0, table_size)
ax.set_ylim(0, table_size)
ax.set_aspect('equal', 'box')

# Plot persegi
square_bowl = plt.Rectangle((0, 0), a, a, linewidth=2, edgecolor='blue', facecolor='none')
ax.add_patch(square_bowl)

# Plot lingkaran
circle_bowl = plt.Circle((a/2, a/2), a/2, linewidth=2, edgecolor='red', facecolor='none')
ax.add_patch(circle_bowl)

# Plot bola yang jatuh dalam persegi tetapi di luar lingkaran
ax.scatter(inside_square_x, inside_square_y, color='blue', alpha=0.5, label='Square Bowl (outside Circle)')

# Plot bola yang jatuh dalam lingkaran
ax.scatter(inside_circle_x, inside_circle_y, color='red', alpha=0.5, label='Circle Bowl')

# Plot bola yang jatuh di luar kedua mangkuk
ax.scatter(outside_bowls_x, outside_bowls_y, color='green', alpha=0.5, label='Outside both Bowls')

# Menambahkan legenda dan label
ax.legend()
ax.set_title("Monte Carlo Simulation: Marble Dropping")
ax.set_xlabel("X Position")
ax.set_ylabel("Y Position")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to simulate the dropping of marbles
def monte_carlo_pi(num_marbles):
    inside_circle = 0
    x_circle = []
    y_circle = []
    x_square = []
    y_square = []
    
    # Simulation loop
    for _ in range(num_marbles):
        # Random point (x, y) within a square with side length 2a (centered at the origin)
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)
        
        # Check if the point is inside the circle (radius = 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1
            x_circle.append(x)
            y_circle.append(y)
        else:
            x_square.append(x)
            y_square.append(y)

    # Estimating Pi as the ratio of marbles in the circle to total marbles
    pi_estimate = 4 * inside_circle / num_marbles
    return pi_estimate, x_circle, y_circle, x_square, y_square

# Number of marbles to drop
num_marbles = 10000

# Run the Monte Carlo simulation
pi_estimate, x_circle, y_circle, x_square, y_square = monte_carlo_pi(num_marbles)

# Visualization
fig, ax = plt.subplots(figsize=(6, 6))

# Plot the square and the circle
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect('equal')

# Plot the points inside the circle
ax.scatter(x_circle, y_circle, color='orange', s=1)
# Plot the points outside the circle (inside the square)
ax.scatter(x_square, y_square, color='blue', s=1)

# Add the circle outline
circle = plt.Circle((0, 0), 1, color='red', fill=False)
ax.add_artist(circle)

# Title and annotations
ax.set_title(f"Monte Carlo Simulation to Estimate Pi\nEstimated Pi: {pi_estimate:.4f}")
plt.show()
```
