```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter distribusi Poisson
lambda_poisson = 5  # Rata-rata jumlah barang cacat per hari
n_simulasi = 30  # Jumlah iterasi simulasi

# Simulasi menggunakan distribusi Poisson
simulasi_poisson = np.random.poisson(lambda_poisson, n_simulasi)

# Visualisasi hasil simulasi
plt.hist(simulasi_poisson, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histogram Jumlah Barang Cacat dalam Produksi Harian")
plt.xlabel("Jumlah Barang Cacat")
plt.ylabel("Frekuensi")
plt.show()

# Menampilkan hasil rata-rata simulasi
rata_rata_simulasi = np.mean(simulasi_poisson)
print(f"Rata-rata jumlah barang cacat dalam simulasi: {rata_rata_simulasi}")
```
