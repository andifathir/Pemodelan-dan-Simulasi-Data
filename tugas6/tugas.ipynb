```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter distribusi Poisson (misal rata-rata jumlah cacat per hari)
lambda_ = 3  # rata-rata barang cacat per hari

# Jumlah iterasi simulasi
iterations = 60

# Simulasi Poisson
simulated_data = np.random.poisson(lam=lambda_, size=iterations)

# Visualisasi hasil simulasi (Histogram)
plt.hist(simulated_data, bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title('Distribusi Poisson - Jumlah Barang Cacat')
plt.xlabel('Jumlah Barang Cacat')
plt.ylabel('Frekuensi')
plt.show()

# Tampilkan rata-rata simulasi
print("Rata-rata simulasi: ", np.mean(simulated_data))
```
